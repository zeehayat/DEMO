```python
import pandas as pd


# Read the CSV file into a DataFrame
df_settlement_data = pd.read_csv('DATA/infrastructure_deliverables.csv')

# Display the first 5 rows
print(df_settlement_data.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
df_settlement_data.info()
```

```python
# One-hot encode categorical variables
data= df_settlement_data
# Convert categorical data into numerical where appropriate for correlation analysis
# For now, we'll focus on numeric fields like 'Completion Percentage' and 'Project Budget (KES)'

# Select numerical columns
numerical_data = data[['Completion Percentage', 'Project Budget (KES)', 'Need New Infrastructure']]

# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()

# Extract correlations that are below 1 but above a certain threshold (e.g., 0.3 to focus on significant correlations)
significant_correlations = correlation_matrix[(correlation_matrix < 1) & (correlation_matrix.abs() > 0.3)]

#significant_correlations

encoded_data = pd.get_dummies(data[['Project Type', 'Project Status', 'Infrastructure Level', 'Population Density', 'Growth Rate']], drop_first=True)

# Combine with the numerical data
combined_data = pd.concat([numerical_data, encoded_data], axis=1)

# Calculate the correlation matrix again with the encoded categorical data
correlation_matrix_encoded = combined_data.corr()

# Extract significant correlations that are below 1 but greater than 0.3 (or less than -0.3 for negative correlations)
significant_correlations_encoded = correlation_matrix_encoded[(correlation_matrix_encoded < 1) & (correlation_matrix_encoded.abs() > 0.3)]

significant_correlations_encoded
```

```python
def clean_correlation_matrix(correlation_matrix):
    # Remove rows and columns where all values are NaN
    cleaned_matrix = correlation_matrix.dropna(how='all').dropna(axis=1, how='all')
    return cleaned_matrix
```

```python
clean_correlation_matrix(significant_correlations_encoded)
```
